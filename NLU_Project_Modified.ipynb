```python
import os
import torch
import transformers
from trl import SFTTrainer
from peft import LoraConfig
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import wandb
import spacy
import matplotlib.pyplot as plt
from google.colab import userdata

```

```python
# Set up environment
os.environ["HF_TOKEN"] = userdata.get('HF_TOKEN')  # Ensure HF_TOKEN is set in Colab secrets
os.environ["WANDB_DISABLED"] = "false"

# Model and quantization configuration
model_id = "google/gemma-2b"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
tokenizer = AutoTokenizer.from_pretrained(model_id, token=os.environ['HF_TOKEN'])
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map={"": 0},
    token=os.environ['HF_TOKEN']
)

```

```python
# Load and split dataset
from datasets import load_dataset
data = load_dataset("bonadossou/instructed_mimic_iii_diagnosis")
data = data["train"].train_test_split(test_size=0.1)
train_data = data["train"]
val_data = data["test"]

```

```python
# prompt: print samples from train data

print(train_data[:1])
```

```python
# Formatting function matching dataset example
def formatting_func(example):
    text = f"{example['instruction_input']}\nAnswer: The patient is high likely exposed to {example['target_disease']}"
    return text  # Return the formatted string directly

# Generate disease prediction
def generate_disease(model, instruction_input, max_new_tokens=20):
    input_text = f"{instruction_input}\nAnswer: The patient is high likely exposed to"
    inputs = tokenizer(input_text, return_tensors="pt").to("cuda:0")
    outputs = model.generate(**inputs, max_new_tokens=max_new_tokens)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

# Extract disease as named entity and plot
def extract_and_plot_disease(response, ground_truth=None):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(response)
    diseases = ["malaria", "tuberculosis", "dengue", "yellow fever"]
    predicted_disease = None

    # Extract disease entity
    for token in doc:
        if token.text.lower() in diseases:
            predicted_disease = token.text.lower()
            break

    # Simple confidence simulation (count-based for demo)
    disease_counts = {disease: 1 if disease == predicted_disease else 0 for disease in diseases}
    if ground_truth:
        print(f"Ground Truth: {ground_truth}")
    print(f"Predicted Disease: {predicted_disease}")

    # Plot bar graph
    plt.figure(figsize=(8, 6))
    plt.bar(disease_counts.keys(), disease_counts.values(), color='skyblue')
    plt.title("Disease Prediction Confidence")
    plt.xlabel("Disease")
    plt.ylabel("Count (Simulated Confidence)")
    plt.savefig("disease_prediction_graph.png")
    plt.close()
    return predicted_disease


```

```python
# Evaluate model
def evaluate_model(model, val_data, num_samples=5):
    correct = 0
    for i in range(min(num_samples, len(val_data))):
        instruction_input = val_data[i]["instruction_input"]
        ground_truth = val_data[i]["target_disease"].lower()
        response = generate_disease(model, instruction_input)
        predicted_disease = extract_and_plot_disease(response, ground_truth if i < 3 else None)
        if predicted_disease == ground_truth:
            correct += 1
        if i < 3:  # Show first 3 examples
            print(f"Input (truncated): {instruction_input[:100]}...")
            print(f"Full Response: {response}\n")
    accuracy = correct / num_samples
    return accuracy

```

```python
# Base model evaluation
print("Base Model Evaluation:")
base_accuracy = evaluate_model(model, val_data)
print(f"Accuracy (Base Model): {base_accuracy}\n")
```

```python
# LoRA configuration
lora_config = LoraConfig(
    r=16,
    target_modules=["q_proj", "o_proj", "k_proj", "v_proj", "gate_proj", "up_proj", "down_proj"],
    task_type="CAUSAL_LM",
)

# Set up trainer
trainer = SFTTrainer(
    model=model,
    train_dataset=train_data,
    args=transformers.TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        warmup_steps=2,
        max_steps=300,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=10,
        output_dir="outputs",
        optim="paged_adamw_8bit"
    ),
    peft_config=lora_config,
    formatting_func=formatting_func,
)

```

```python
# Initialize Weights & Biases
wandb.init(project="Disease-v1-Prediction", name="instructed_mimic_iii_run")

# Train the model
trainer.train()
```

```python

# Fine-tuned model evaluation
print("Fine-Tuned Model Evaluation:")
fine_tuned_accuracy = evaluate_model(model, val_data)
print(f"Accuracy (Fine-Tuned Model): {fine_tuned_accuracy}\n")

```
